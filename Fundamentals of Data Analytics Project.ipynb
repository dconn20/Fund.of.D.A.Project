### Linear Regression

Linear regression is one of the fundamental statistical and machine learning techniques. It is a statistical approach for modelling the relationship between a dependent variable with a given set of independent variables based on supervised learning.  It is mostly used for finding out the relationship between variables and forecasting<sup>7</sup>. The model will be created using the training data. The accuracy of the model can then be evaluated using different metrics such as: 

* Mean Absolute Error: This measures the average distance between the actual values and the predicted values.<sup>3</sup>


* Mean Squared Error: The measure of the difference between the two continuous variables. The Mean Squared Error is the average vertical distance between each actual value and the line that best matches the data.<sup>3</sup>  


* Root Mean Squared Error: This is defined as the square root of the average squared distance between the actual score and the predicted score<sup>3</sup> 


* R-squared: It is also known as coefficient of determination and it is used to assess how close the data is to the regression line or how well the model has fit the data. R² is always between 0 and 1 or between 0% to 100%.<sup>3</sup>
<br/>

#### Perform Linear Regession

Use Scikit-learn (sklearn) to create a machine learning model that performs simple linear regression on the speed and power data. The model should be able to take in a new wind speed measurement and accurately predict the power output from a wind turbine. A good fitting model is one where the difference between the actual and observed values or predicted values for the selected model is small and unbiased for training and test data sets<sup>3</sup>. In order for the data to be used in machine learning algorithms it will need to be converted into a numpy array<sup>4</sup>. The data will then be seperated into training data and test data. The algorithm uses the training data to learn and the test data to evaluate the performance of the model. The test data will contain 30% of the dataset, meaning 315 rows will be used to train the model while 136 rows will be used for testing purposes.  

*******************************************

#### Prepare Dataset 

```python
# Convert the data from a pandas dataframe to a numpy array for use in the models [5]

x = df['speed'].values.reshape(-1,1)
y = df['power'].values.reshape(-1,1)

# Split the dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state=0)

# Print size of each set
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

**********************************************************************
#### View the training and testing data

```python
# Plot the training and testing data [6]

# Set the parameters for the subplots to be created
fig, axes = plt.subplots(1, 2, figsize=(16,8))

# Plot the data set before and after cleaning for comparison purposes
axes[0].plot(x_train, y_train, 'b.')
axes[1].plot(x_test, y_test, 'r.')

# Add labels 
axes[0].set_xlabel('Speed', fontsize=18)
axes[0].set_ylabel('Power', fontsize=18)
axes[1].set_xlabel('Speed', fontsize=18)
axes[1].set_ylabel('Power', fontsize=18)

# Add titles to subplots
axes[0].set_title('Training Set', fontsize=18)
axes[1].set_title('Test set', fontsize=18)
```

****************************************

#### Build and train the model

```python
# Creating the model [3]
lin_model = LinearRegression()

# Training the model with training data
lin_model.fit(x_train, y_train)
```

*************************************

#### Evaluate the model

```python
# Get R-squared value [3]
r2 = lin_model.score(x,y)
r2
```

```python
# Get the linear regression intercepts [3]
p = [lin_model.intercept_, lin_model.coef_[0]]
print(p)
```

```python
# Use predict to test the model
lin_pred = lin_model.predict(x_test)
```

```python
# Compare predictions against acutal values [8] 
# Use flatten() to return a copy of the array collapsed into one dimension [8]

testdata = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': lin_pred.flatten()})
testdata
```

```python
# Plot predictions against the data [9]
plt.plot(x_test, y_test, 'r.', label='actual')
plt.plot(x_test, lin_pred, 'b-', label='prediction')
plt.title("Dataset versus Predictions", fontsize = 18)
plt.xlabel("Wind Speed")
plt.ylabel("Power Output")
plt.legend();
```

*********************************************************************

#### Conclusion

Although the coefficient of determination also known as the R-squared value is quite high at almost 89 it is clear to see from the created dataset that there is a large variation between the predicted values and the actual values. The graph above also shows a variation in the values as the linear predictions are in a straight line while the actual data has an S shape. This model does not seem to replicate the relationship between speed and power accurately.

### Polynomial Regression

Polynomial regression is a form of linear regression in which the best fit line is not a straight line instead it is in the form of a curve. Polynomial regression would seem to fit the data better than linear regression as it follows the flow of the actual data much more closely. We can choose the degree of polynomial based on the relationship between target and predictor. The 1-degree polynomial is a simple linear regression so the value of degree must be greater than 1<sup>10</sup>. For this model the value of degree will be set to 4 to try match the data.

*******************************************

#### Build and train the model

```python
#Fitting the polynomial regression model to the dataset [11]

poly_reg=PolynomialFeatures(degree=4)

x_trainpoly=poly_reg.fit_transform(x_train)
polyreg=LinearRegression()
polyreg.fit(x_trainpoly, y_train)
x_test_poly = poly_reg.fit_transform(x_test)
```

*********************************************************

#### Evaluate the model

```python
# Get R-squared value [3]
p_r2 = polyreg.score(x_test_poly,y_test)
p_r2
```

```python
# Get the polynomial regression intercepts [3]
polyp =[[polyreg.intercept_, lin_model.coef_[0]]]
polyp
```

```python
# Use predict to test the model
poly_pred = polyreg.predict(poly_reg.fit_transform(x_test))
```

```python
# Compare predictions against acutal values [8]
testdata2 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': poly_pred.flatten()})
testdata2
```

### Compare The Models
